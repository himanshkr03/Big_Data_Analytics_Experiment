```python
# NAME: HIMANSAHU RAJAK
# UID : 21BCS1198
import pandas as pd
import numpy as np
def analyze_big_data(dataset_path):
    # Step 1: Load the dataset
    data = pd.read_csv("/content/phone_usage_india.csv")
    # Step 2: Volume - Count number of rows and columns
    volume = data.shape  # (rows, columns)
    # Step 3: Variety - Check data types and unique values
    variety = data.dtypes
    unique_values = data.nunique()
    # Step 4: Velocity - Estimate update frequency (if time-based data available)
    if 'timestamp' in data.columns:
        data['timestamp'] = pd.to_datetime(data['timestamp'])
        velocity = data['timestamp'].diff().mean()  # Average time between entries
    else:
        velocity = "Not Applicable"
    # Step 5: Veracity - Check for missing values or inconsistencies
    missing_data = data.isnull().sum()
    # Step 6: Value - Perform basic analysis (e.g., mean, median)
    value = data.describe()  # Summary statistics
    # Step 7: Return all 5 V's results
    return volume, variety, unique_values, velocity, missing_data, value
# Example usage
dataset_path = 'your_dataset.csv'
volume, variety, unique_values, velocity, missing_data, value = analyze_big_data(dataset_path)
# Output results
print("-----Volumn---:")
print(volume)
print("-----Variety--:")
print(variety)
print("-------Unique Values--:")
print(unique_values)
print("-------Velocity-")
print(velocity)
print("------Missing Data--")
print(missing_data)
print("----Value Analysis--")
print(value)
```
